#Importando o Pandas e o Matplotlib


```python
import pandas as pd
import matplotlib.pyplot as plt
```

#importando os dados

```python
df = pd.read_csv('/content/dados_rh.csv')
```

#Olhando o banco

```python
df
```

#O banco tem 35 colunas e 1470 linha, não tem dados do tipo nulo, tem 26 dados do tipo int e 9 do tipo object

```python
df.info()
```

```python
df2 = df.groupby('Department')[['EmployeeNumber']].count()
```

#Quantidade de funcionários por setor, sendo 63 em Human Resources, 961 em Research & Development e 446 em Sales.

```python
df2
```

```python
df2.plot(kind = 'bar')
```

```python
df3 = df.groupby('Education')[['EmployeeNumber']].count()
```

#Quantidade de funcionários por nível de educação, no nível 1 temos 170 funcionários, no 2 temos 282, no 3 temos 572, no 4 temos 398 e no 5 temos 48 funcionários, sendo assim a maioria dos funcionários tem uma educação de nível 3.

```python
df3
```

```python
df3.plot(kind='line',subplots=True)
```

```python
df4 = df.groupby('Attrition')[['EmployeeNumber']].count()
```

#Quantidade de funcionários que tem ou não algum atrito na empresa, sendo 1233 que não tem atrito e 237 que tem algum tipo de atrito, a maior parte dos funcionários não tem nem um tipo de atrito.


```python
df4
```

```python
df4.plot(kind='pie',subplots=True)
```
